```python
movies = pd.read_csv("/kaggle/input/movie-rating-dataset/movie.csv")
movies.head()
```

```python
ratings = pd.read_csv("/kaggle/input/movie-rating-dataset/rating.csv")
ratings.head()
```

```python
df = ratings.merge(movies, how = "left",on = "movieId")
df.head()
```

```python
df_filtered = df.groupby("movieId").agg({"rating":"count"})
df_filtered.head()
```

```python
type(df_filtered)
```

```python
df_filtered = df_filtered[df_filtered["rating"] > 1000]
df_filtered.sort_values(by = "rating",ascending=True)
```

```python
df_filtered = df[df["movieId"].isin(df_filtered.index.to_list())]
df_filtered.head()
```

```python
df_filtered.shape
```

```python
df_pivot = df_filtered.pivot_table(values="rating",index="userId",columns="title")
df_pivot.head()
```

```python
random_user = df_filtered["userId"].sample(1,random_state=45).values[0]
random_user
```

```python
random_user_liked_movie = df_filtered.loc[(df_filtered["userId"] == random_user) & (df_filtered["rating"] == 5.0)  ]
random_user_liked_movie
```

```python
random_user_liked_movie["timestamp"] =pd.to_datetime( random_user_liked_movie["timestamp"])
```

```python
random_user_liked_movie["timestamp"]
```

```python
random_user_liked_movie.sort_values(by="timestamp",ascending=False)
```

```python
last_movie = random_user_liked_movie.sort_values(by="timestamp",ascending=False).iloc[0]["title"]
last_movie
```

```python
last_movie_column = df_pivot[last_movie]
last_movie_column
```

```python
df_pivot.corrwith(last_movie_column).sort_values(ascending=False).head(10)
```
